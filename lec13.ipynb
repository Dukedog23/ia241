```python
import pandas
```

```python
df=pandas.read_excel('s3://zeng-ia241-2021spring/house_price.xls')

df.sort_values(by=['built_in']).plot(x='built_in',y='price')
```

```python
avg_price_per_year=df.groupby('built_in').mean()['price']
```

```python
avg_price_per_year.plot()
```

```python
df['price'].hist()
```

```python
avg_price_per_type=df.groupby('house_type').mean()['price']
avg_price_per_type[:]
```

```python
avg_price_per_type.plot.bar()
```

```python
count_per_type=df['house_type'].value_counts()
count_per_type[:]
```
